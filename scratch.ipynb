```python
# import libraries for reading and parsing RSS feeds
import feedparser

# Simple XML parser for reading the list of feeds in feeds.xml
import xml.etree.ElementTree as ET
```

```python
# Load all of the fees in feeds.xml

tree = ET.parse('feeds.xml')

# Explore the tree
root = tree.getroot()
print(root.tag)
print(root.attrib)

# Iterate over the children of the root
for child in root:
    print(child.tag, child.attrib)
    for grandchild in child:
        print(grandchild.tag, grandchild.attrib)
```
